#### Step 2: Creating a K-Means model with K=2, and then fitting it

```python
kmeans_model = KMeans(n_clusters=2, random_state=0)#creates a k-means model with skit-learn's kmeans
```

```python
fitted_model = kmeans_model.fit(cluster_data)#fits the model to our data
```

```python
fitted_model.labels_ #these are the predictions
```

```python
cluster_data['2Means Cluster Predictions'] = fitted_model.labels_
```

#### Step 4: Inspecting this K-Means model

```python
### Let's look at the first 10 rows
cluster_data.head(n=10)
```

```python
cluster0 = cluster_data[cluster_data['2Means Cluster Predictions'] == 0] #only getting cluster data where this 'cluster_data['2Means Cluster Predictions'] == 0' is true
cluster1 = cluster_data[cluster_data['2Means Cluster Predictions'] == 1]
```

```python
#find the means for all the columns
for column in cluster_data.columns:
    mean0 = cluster0[column].mean()
    mean1 = cluster1[column].mean()
    print("")
    print("Evaluating column: **{c}**".format(c=column))
    print("        Cluster 0 Mean: {m0:.3f}, Cluster 1: {m1:.3f}".format(m0=mean0, m1=mean1))
    print("        Difference in means = {diff:.3f}".format(diff=abs(mean0 - mean1)))
    print("")
    print("-----------------------------------------------------------------")
    
    
```

#### Step 3: Replicating the with DB Scan

```python
from sklearn.preprocessing import StandardScaler 
```

```python
X = StandardScaler().fit_transform(cluster_data) #scale all data, so they are on the same scale -
#just a data processing technique
```

```python
pd.DataFrame(data=X)
```

Originally, eps = 3, min_samples = 3, but that didn't work for my dataset, because it only estimated 1 cluster, with a Silhouette coefficient of 0.672.

I noticed that as eps increased, the number of clusters predicted tended to decrease, while changing min_samples did not have a large effect

I feel that a more accurate dbscan_model has eps=2, min_samples=2, because although the Silhouette coefficient decreased (to .370), it predicted 6 clusters, which seems more reasonable (as having 1 cluster really doesn't mean the model clustered anything). Secondly, I know from the dataset that the data is comparing the average prices spent on various grocery items from 3 regions and 2 channels, so it is highly likely that there would be 6 different clusters separating the data.  

However, when I changed it to eps = 3, keeping min_samples at 2, I got an estimated number of clusters as 2, with a higher silhouette coefficient (.619). This makes sense because the data is split between two channels, and it could be that the different regions themselves are not that important, rather the deciding factor is the channels. 

```python
dbscan_model = DBSCAN(eps=3, min_samples=2) 
```

```python
fitted_model = dbscan_model.fit(X)
```

```python
labels = fitted_model.labels_ #getting the predictions, like above with Kmeans
```

```python
labels
```

#### Step 4: Using sklearn metrics to measure goodness of fit

```python
from sklearn import metrics
```

```python
# #############################################################################
# Compute DBSCAN Metrics

# Number of clusters in labels, ignoring noise if present.
if -1 in labels:
    n_clusters_ = len(set(labels)) - 1
else:
    n_clusters = len(set(labels))

print('Estimated number of clusters: {n}'.format(n=n_clusters_))
print("Silhouette Coefficient: {:0.3f}".format(metrics.silhouette_score(X, labels)))
```
